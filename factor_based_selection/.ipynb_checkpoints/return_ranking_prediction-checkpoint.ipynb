```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xgboost
from xgboost import XGBClassifier
from xgboost import XGBRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Data loading
data = pd.read_csv('./数据汇总-超额收益率.csv',sep=',')

# Adding an additional column as the ranking of the return
data['rank'] = data.groupby(['year','trade_date'])['label'].rank(method='min')
count = data.groupby(['year','trade_date']).size().reset_index(name='count')
data = data.join(count.set_index(['year','trade_date']), on=['year','trade_date'])
data['rank'] = data['rank']/data['count']

# Extracting a list of years and trade dates
year = data[['year']].drop_duplicates(subset=None, keep='first', inplace=False)
year = (np.array(year)).tolist() 
trade_date = data[['trade_date']].drop_duplicates(subset=None, keep='first', inplace=False)
trade_date = (np.array(trade_date)).tolist()

# Init feature importance list
feature_importance = []
```

```python
# 2006-2009训练，2010验证调参，预测2011
#yy为起始年份，yy-yy+3训练，yy+4验证调参，预测yy+5
#yy = 2006 - 2017
yy =  2006#请输入
```

```python
X_train = data[data.year.isin([yy,yy+1,yy+2,yy+3])].iloc[:,4:45]
X_validate = data[data.year.isin([yy+4])].iloc[:,4:45]
X_test = data[data.year.isin([yy+5])].iloc[:,4:45]
Y_train = data[data.year.isin([yy,yy+1,yy+2,yy+3])][['rank']].values
Y_validate = data[data.year.isin([yy+4])][['rank']].values
Y_test = data[data.year.isin([yy+5])][['rank']].values
```

```python
#模型5  2010-2013训练，2014验证调参，预测2015
scaler = StandardScaler()
scaler.fit(X_train)

model = XGBRegressor(max_depth=6,                       #这个值为树的最大深度       
                     n_estimators=2000,                     #最大的迭代次数           
                     gamma=3,                 
                     min_child_weight=1,      
                     subsample=.95,
                     learning_rate=.05)
#                      colsample_bylevel=1
#                      objective='reg:logistic',   
#                      booster='gblinear',
#                      random_state=0)           
model.fit(scaler.transform(X_train),Y_train)

Y_train_pred = model.predict(scaler.transform(X_train))
Y_validate_pred = model.predict(scaler.transform(X_validate))
Y_test_pred = model.predict(scaler.transform(X_test))


print(r2_score(Y_train, Y_train_pred))
print(r2_score(Y_validate, Y_validate_pred))
print(r2_score(Y_test, Y_test_pred))
```

```python
from sklearn.linear_model import LinearRegression
##日频收益率预测
# True rank of the return rate is sorted and ploted in orange
# Predicted rank ploted in blue
# Top one-fifth predictions are marked with red dots
# Trend of the prediction is ploted in red line

def plot_result(idx, Y_vis, Y_vis_pred, name):
    plt.figure(figsize = (24,6))
    plt.plot(Y_vis_pred[idx])
    plt.plot(Y_vis[idx])
    model = LinearRegression().fit(np.arange(500).reshape(-1, 1),Y_vis_pred[idx])
    y = model.predict([[0],[499]])
    plt.plot([0,499], y,'r')
    selection = np.argsort(Y_vis_pred[idx])[-100:]
    plt.plot(selection, np.ones(len(selection))*.75,'r.')
    plt.legend(['pred', 'true'])
    plt.title(name)
    
    
experiments = [ \
    {'name':'train1', 'idx':np.argsort(Y_train[:500,0]), 'Y':Y_train, 'Y_pred':Y_train_pred}, \
    {'name':'train2', 'idx':np.argsort(Y_train[10500:11000,0])+10500, 'Y':Y_train, 'Y_pred':Y_train_pred}, \
    {'name':'val1', 'idx':np.argsort(Y_validate[:500,0]), 'Y':Y_validate, 'Y_pred':Y_validate_pred}, \
    {'name':'val2', 'idx':np.argsort(Y_validate[10500:11000,0])+10500, 'Y':Y_validate, 'Y_pred':Y_validate_pred}, \
    {'name':'test1', 'idx':np.argsort(Y_test[:500,0]), 'Y':Y_test, 'Y_pred':Y_test_pred}, \
    {'name':'test2', 'idx':np.argsort(Y_test[10500:11000,0])+10500, 'Y':Y_test, 'Y_pred':Y_test_pred}, \
    ]

for e in experiments:
    plot_result(e['idx'],e['Y'],e['Y_pred'],e['name'])
    

print(r2_score(Y_train, Y_train_pred))
print(r2_score(Y_validate, Y_validate_pred))
print(r2_score(Y_test, Y_test_pred))
```

```python
importance = model.feature_importances_
feature_importance.append(importance)
print(importance)
```

```python
feature_importance_avg = np.mean(feature_importance, axis=0)
```

```python
plot_order = np.argsort(feature_importance_avg)
plt.figure(figsize=(10,10))
plt.barh(np.arange(len(feature_importance_avg)), feature_importance_avg[plot_order], \
    tick_label=data.columns[4:45][plot_order], align='center')
plt.show()
```
