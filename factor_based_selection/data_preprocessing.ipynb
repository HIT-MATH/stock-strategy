```python
import os
import pandas as pd
import numpy as np
```

```python
# generate file list
file_path_101 = [os.path.join("./101neu", p) for p in np.sort(os.listdir("./101neu"))]
file_path_191 = [os.path.join("./191neu", p) for p in np.sort(os.listdir("./191neu"))]
file_path_360 = [os.path.join("./360neu", p) for p in np.sort(os.listdir("./360neu"))[:240]]
file_path = np.concatenate([file_path_101, file_path_191, file_path_360])
file_path = np.delete(file_path, np.where(file_path=='./191neu/alpha191-056.pkl'))
file_path = np.delete(file_path, np.where(file_path=='./101neu/alpha101-061.pkl'))
file_path = np.delete(file_path, np.where(file_path=='./101neu/alpha101-075.pkl'))
file_path = np.delete(file_path, np.where(file_path=='./101neu/alpha101-095.pkl'))
file_path = np.delete(file_path, np.where(file_path=='./191neu/alpha191-154.pkl'))
```

```python
print(file_path, len(file_path), len(file_path_101), len(file_path_191))
```

```python
date = pd.read_csv("data.csv",sep=',')
date_list = date.index[date["trade_date"]>20100000].to_list()
```

```python
df = date.iloc[date_list]
for p in file_path:
    print(p)
    df_new = pd.read_pickle(p).iloc[date_list]
    df = pd.concat([df, df_new],axis=1)
```

```python
df["year"] = (df['trade_date']/(1e4)).astype(int)
df = df.reset_index()
```
